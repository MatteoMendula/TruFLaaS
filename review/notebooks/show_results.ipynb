```python
import numpy as np
acc = [0.3658, 0.505, 0.4468, 0.4863, 0.581, 0.4937, 0.3694, 0.5066, 0.4359, 0.7534]

mean = np.mean(acc)
std = np.std(acc)

selected_truflaas = {}
selected_trustfed = {}

for index, a in enumerate(acc):
    if (a > mean - 0.2 * std):
        selected_truflaas[index] = a
    if (a > mean - 0.2 * std and a < mean + 0.2 * std):
        selected_trustfed[index] = a
        
print("selected_truflaas", selected_truflaas)
print("selected_trustfed", selected_trustfed)
```
